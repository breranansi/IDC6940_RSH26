```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bc=-0.3624
v=np.linspace(70,1670,50)
v_bc=(np.pow(v,bc)-1.0)/bc
v_ln=np.log(v)
v_sq=1.0/np.sqrt(v)
```

```python
#plt.scatter(v,v_ln,marker="o",c='red',edgecolor="black",label="log")
plt.scatter(v,v_bc,marker="v",c='green',edgecolor="black",label="box-cox")
plt.scatter(v,v_sq,marker="s",c='blue',edgecolor="black",label="inverse-sqrt")
plt.xlabel("Raw Molecular Mass")
plt.ylabel("Transformed Molecular Mass")
plt.grid()
plt.legend()
plt.show()
```
