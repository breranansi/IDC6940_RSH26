```python
# Call libraries
import numpy as np
import pandas as pd
import sympy as sym
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import time
```

```python
# Build Toy Dataset
x=np.arange(start=0.55,stop=1.075,step=0.025)
x0=np.log(x)
x1=np.exp(x)
x2=x**2
x3=x**3
x4=1/x
x5={'x':x,
    'log':x0,
    'exp':x1,
    'sqr':x2,
    'cub':x3,
    'inv':x4}
df=pd.DataFrame(x5)
y=141.5/x-131.5
plt.plot(x,y,marker="o",linestyle="None")
plt.show()
```

```python
t0 = time.time()
myMod=LinearRegression()
myMod.fit(df,y)
y_pred=myMod.predict(df)
print("time to fit: ",time.time() - t0, 'seconds')
```

```python
plt.plot(x,y,marker="o",linestyle="None")
plt.plot(x,y_pred)
plt.show()
```

```python
print(f"x:   {myMod.coef_[0]:.4f}")
print(f"log: {myMod.coef_[1]:.4f}")
print(f"exp: {myMod.coef_[2]:.4f}")
print(f"sqr: {myMod.coef_[3]:.4f}")
print(f"cub: {myMod.coef_[4]:.4f}")
print(f"inv  {myMod.coef_[5]:.4f}")
print(f"int: {myMod.intercept_:.4f}")
```

```python
dfd=df.drop(columns=['inv'])
myNuMod=LinearRegression()
myNuMod.fit(dfd,y)
y_new=myNuMod.predict(dfd)
plt.plot(x,y,marker="o",linestyle="None")
plt.plot(x,y_new)
plt.show()
```

```python
print(f"x:   {myNuMod.coef_[0]:.4f}")
print(f"log: {myNuMod.coef_[1]:.4f}")
print(f"exp: {myNuMod.coef_[2]:.4f}")
print(f"sqr: {myNuMod.coef_[3]:.4f}")
print(f"cub: {myNuMod.coef_[4]:.4f}")
print(f"int: {myNuMod.intercept_:.4f}")
```

```python
from sklearn.linear_model import Ridge
myRidge=Ridge(alpha=0.5)
myRidge.fit(df,y)
y_ridge=myRidge.predict(df)
plt.plot(x,y,marker="o",linestyle="None")
plt.plot(x,y_ridge)
plt.show()
```

```python
print(f"x:   {myRidge.coef_[0]:.4f}")
print(f"log: {myRidge.coef_[1]:.4f}")
print(f"exp: {myRidge.coef_[2]:.4f}")
print(f"sqr: {myRidge.coef_[3]:.4f}")
print(f"cub: {myRidge.coef_[4]:.4f}")
print(f"int: {myRidge.intercept_:.4f}")
```

```python
from sklearn.linear_model import Lasso
myLasso=Lasso(alpha=0.5)
myLasso.fit(df,y)
y_lasso=myLasso.predict(df)
plt.plot(x,y,marker="o",linestyle="None")
plt.plot(x,y_lasso)
plt.show()
```

```python
print(f"x:   {myLasso.coef_[0]:.4f}")
print(f"log: {myLasso.coef_[1]:.4f}")
print(f"exp: {myLasso.coef_[2]:.4f}")
print(f"sqr: {myLasso.coef_[3]:.4f}")
print(f"cub: {myLasso.coef_[4]:.4f}")
print(f"int: {myLasso.intercept_:.4f}")
```
