```python
# Call libraries
import numpy as np
import pandas as pd
import sympy as sym
import matplotlib.pyplot as plt
from pyoperon.sklearn import SymbolicRegressor
import time
```

```python
# Build Toy Dataset
x=np.arange(start=0.55,stop=1.075,step=0.025)
y=141.5/x-131.5
plt.plot(x,y,marker="o",linestyle="None")
plt.show()
```

```python
t0 = time.time()
myMod=SymbolicRegressor(random_state=7)
myMod.fit(x.reshape(-1, 1),y)
y_pred=myMod.predict(x.reshape(-1, 1))
print("time to fit: ",time.time() - t0, 'seconds')
```

```python
plt.plot(x,y,marker="o",linestyle="None")
plt.plot(x,y_pred)
plt.show()
```

```python
myEq=myMod.sympy()
myEq
```

```python
sym.simplify(myEq)
```
